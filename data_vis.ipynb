```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age_moyen'], bins=20, kde=True)
plt.title('Distribution de l\'âge moyen par commune')
plt.xlabel('Âge moyen')
plt.ylabel('Fréquence')
plt.grid(True)
plt.show()
```

```python
data_unique = df.drop_duplicates(subset=['libelle_commune', 'catégorie', 'nom', 'voix'])
pivot_data = data_unique.pivot_table(index='nom', columns='catégorie', values='voix', aggfunc='sum').fillna(0)

# Création de sous-tracés (subplots) pour chaque catégorie
fig, axes = plt.subplots(len(pivot_data.columns), 1, figsize=(8, 6 * len(pivot_data.columns)))

# Création d'un graphique "bar of pie" pour chaque catégorie
for i, categorie in enumerate(pivot_data.columns):
    # Données pour le graphique
    noms = pivot_data.index
    valeurs = pivot_data[categorie]
    
    # Création du graphique circulaire
    axes[i].pie(valeurs, labels=noms, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(categorie)  # Titre du sous-tracé

# Réglages de l'espacement entre les sous-tracés
plt.tight_layout()

# Affichage du graphique
plt.show()
```

```python
import numpy as np
voix_par_circo_par_candidat = df.groupby(['circo', 'nom'])['voix'].sum().reset_index()

voix_par_circo = df.groupby('circo')['voix'].sum().reset_index()
voix_par_circo.rename(columns={'voix': 'total_nb_de_par_ciro'}, inplace=True)

voix_par_circo_par_candidat_pourcentage = pd.merge(voix_par_circo_par_candidat, voix_par_circo, on='circo')

voix_par_circo_par_candidat_pourcentage['pourcentage_voix'] = (voix_par_circo_par_candidat_pourcentage['voix'] / voix_par_circo_par_candidat_pourcentage['total_nb_de_par_ciro']) * 100

voix_par_circo_par_candidat_pourcentage = pd.merge(voix_par_circo_par_candidat_pourcentage, df[['circo', 'nom', 'part_pauvres_diff','part_modestes_diff','part_medians_diff','part_plutot_aises_diff','part_aises_diff','actemp','actcho','inactret','age_moyen']], on=['circo', 'nom'])

indice_max_par_circo = voix_par_circo_par_candidat_pourcentage.groupby('circo')['pourcentage_voix'].idxmax()

pourcentage_max_par_circo = voix_par_circo_par_candidat_pourcentage.loc[indice_max_par_circo, ['circo', 'nom', 'pourcentage_voix', 'part_pauvres_diff','part_modestes_diff','part_medians_diff','part_plutot_aises_diff','part_aises_diff','actemp','actcho','inactret','age_moyen']]

float_columns = pourcentage_max_par_circo.select_dtypes(include=float).columns
float_columns = [col for col in float_columns if col != 'pourcentage_voix']

for column in float_columns:
    top_15 = pourcentage_max_par_circo.sort_values(by=column, ascending=False).head(15)
    plt.figure(figsize=(10, 6))
    
    bar_width = 0.35
    indices = np.arange(len(top_15))
    
    plt.bar(indices, top_15[column], bar_width, color='skyblue', label=column)
    
    plt.xlabel('Nom des candidats')
    plt.ylabel(column)
    plt.title(f'{column} des 15 premiers candidats avec les pourcentages de voix maximum')
    plt.xticks(indices, top_15['nom'], rotation=45, ha='right') 
    plt.tight_layout()
    
    plt.show()
```

```python
# Différences des parts de populations par circonscription
population_diff_columns = ['part_pauvres_diff', 'part_modestes_diff', 'part_medians_diff', 
                           'part_plutot_aises_diff', 'part_aises_diff']
data_melted = pourcentage_max_par_circo.melt(id_vars=['circo'], value_vars=population_diff_columns, 
                        var_name='Population_Type', value_name='Difference')

plt.figure(figsize=(14, 8))
sns.boxplot(x='Population_Type', y='Difference', data=data_melted)
plt.title('Différences des parts de populations par circonscription')
plt.xticks(rotation=45)
plt.show()
```

```python
float_columns = pourcentage_max_par_circo.select_dtypes(include=float).columns
float_columns = [col for col in float_columns if col != 'pourcentage_voix']

for column in float_columns:
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x=column, y='pourcentage_voix', hue='nom', data=pourcentage_max_par_circo, palette='tab10')
    plt.title('Relation entre pourcentage de voix et part de pauvres')
    plt.show()
```
