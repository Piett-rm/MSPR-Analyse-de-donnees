```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('./all_donnees.csv')
```

```python
df
```

```python
colonnes_num = df.select_dtypes(include=['float']).columns
print(colonnes_num)
```

```python
df.describe()
```

```python
for colonne in colonnes_num:
    plt.figure(figsize=(10,6))
    plt.title(f"Histogramme de {colonne}")
    sns.histplot(df, x=colonne, kde=True)
```

```python
for column in colonnes_num:
    plt.figure(figsize=(10,6))
    plt.title(column)
    sns.boxplot(df, x=df[column])
```

```python
sns.countplot(x='is_genuine', data=df)
```
