```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('./all_donnees.csv')
```

```python
df
```

```python
colonnes_num = df.select_dtypes(include=['float']).columns
print(colonnes_num)
```

```python
df.describe()
```

```python
for colonne in colonnes_num:
    plt.figure(figsize=(10,6))
    plt.title(f"Histogramme de {colonne}")
    sns.histplot(df, x=colonne, kde=True)
```

```python
for column in colonnes_num:
    plt.figure(figsize=(10,6))
    plt.title(column)
    sns.boxplot(df, x=df[column])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age_moyen'], bins=20, kde=True)
plt.title('Distribution de l\'âge moyen par commune')
plt.xlabel('Âge moyen')
plt.ylabel('Fréquence')
plt.grid(True)
plt.show()
```

```python
pivot_data = df.pivot_table(index='nom', columns='catégorie', values='voix', aggfunc='sum').fillna(0)
print(pivot_data)

plt.figure(figsize=(14, 8))

pivot_data.plot(kind='bar', stacked=True, figsize=(14, 8))

plt.title('Répartition des voix obtenues par catégorie avec les noms des participants')
plt.xlabel('Nom des participants')
plt.ylabel('Nombre de voix')
plt.xticks(rotation=90)
plt.legend(title='Catégorie')

plt.tight_layout()
plt.show()
```

```python
data_unique = df.drop_duplicates(subset=['libelle_commune', 'catégorie', 'nom', 'voix'])
pivot_data = data_unique.pivot_table(index='nom', columns='catégorie', values='voix', aggfunc='sum').fillna(0)

plt.figure(figsize=(14, 8))

pivot_data.plot(kind='bar', stacked=True, figsize=(14, 8))

plt.title('Répartition des voix obtenues par catégorie avec les noms des participants')
plt.xlabel('Nom des participants')
plt.ylabel('Nombre de voix')
plt.xticks(rotation=90)
plt.legend(title='Catégorie')

plt.tight_layout()
plt.show()
```

```python
totals = pivot_data.sum(axis=0)

percentages = pivot_data.div(totals, axis=1) * 100
print(percentages)
print((29.66+31.6+25.5+20.9)/4)
```

```python
part_pauvres_diff_per_city = df[['libelle_commune', 'part_pauvres_diff']].drop_duplicates(subset=['libelle_commune'])

top_candidates_per_city = df.loc[df.groupby('libelle_commune')['pourcentage_voix'].idxmax()]

top_candidates_per_city_result = top_candidates_per_city[['libelle_commune', 'nom', 'pourcentage_voix']]


data = part_pauvres_diff_per_city.merge(top_candidates_per_city_result, on='libelle_commune', how="inner")
top_10_pauvres = data.sort_values(by='part_pauvres_diff', ascending=False).head(10)
top_10_pauvres
```
